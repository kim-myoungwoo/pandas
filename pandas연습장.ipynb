```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from urllib import request
from bs4 import BeautifulSoup
```

```python
url='https://new.land.naver.com/houses?ms=37.538617,127.082375,15&a=VL:DDDGG:JWJT:SGJT:HOJT&e=RETAIL'
```

```python
target = request.urlopen(url)

soup = BeautifulSoup(target.read())
soup
```

```python
house_name = []
house_price = []
for line in soup.find_all('div'):
    name = line.find('span', class_='text')
    if name:        
        house_name.append(name.get_text().strip())
        
    price = line.find('span', class_='price')
    if house_price:
        house_price.append(price.get_text())
        
print(house_name)
print(house_price)
```
