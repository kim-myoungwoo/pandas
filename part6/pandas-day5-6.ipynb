```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

plt.rc('font', family='Malgun Gothic')
```

```python
url='https://sports.news.naver.com/kbaseball/record/index?category=kbo&year='
```

```python
df=pd.DataFrame([])
```

```python
df1=pd.read_html(url)[0]
```

```python
df1
```

```python
for i in range(2012,2022+1):
    df1= pd.read_html(url+str(i))[0]
    df1['연도']=str(i)
    df=pd.concat([df,df1])
```

```python
df.head(30)
```

```python
df=df.replace({'kt':'KT','SK':'SSG','넥센':"키움"})
```

pivot(): 두개의 열을 기준으로 행과 열을 재구성
    
pivot_table(): 교재에서 배운 함수, 데이터 집계 함수를 사용

```python
df_pivot1= df.pivot(index="연도",columns="팀", values='순위')
```

```python
df_pivot1
```

```python
df_pivot2=df.pivot(index="연도",columns="순위",values='팀')
```

```python
df_pivot2
```

```python
df_pivot3=df.pivot(index="연도", columns="팀", values="승률")
```

```python
df.pivot3
```

```python
data = {'team': ['LG', '두산', '한화'],
        'marker_color': ['red', 'navy', 'orange'],
        'face_color': ['black', 'yellow', 'white']}
df_three_team=pd.DataFrame(data)
```

```python
df_three_team
```

```python
plt.style.use('ggplot')

fig=plt.figure(figsize=(12,5))
ax1=fig.add_subplot(1,2,1)
ax2=fig.add_subplot(1,2,2)

# 연도별 순위/승률
# LG, 두산, 한화
# ax1 => 팀 순위/ ax2 => 팀 승점

for team, marker_color, face_color in zip(df_three_team['team'],df_three_team['marker_color'],df_three_team['face_color']):
    
    ax1.plot(df_pivot1.index,df_pivot1[team], marker='o', color=marker_color, markerfacecolor=face_color,label=team)
    ax2.plot(df_pivot3.index,df_pivot3[team], marker='o', color= marker_color, markerfacecolor=face_color,label=team)
    
    
    #     print(team, marker_color, face_color)

# ax1.plot(df_pivot1.index, df_pivot1['LG'], marker='o', color='red', markerfacecolor='black')
# ax1.plot(df_pivot1.index, df_pivot1['두산'], marker='o', color='navy', markerfacecolor='yellow')
# ax1.plot(df_pivot1.index, df_pivot1['한화'], marker='o', color='yellow', markerfacecolor='white')



ax1.set_ylim(10.5,0.5)
# ax2.set_ylim(0.2,0.8)
ax2.legend(bbox_to_anchor=(1,1))
ax1.legend(bbox_to_anchor=(-0.05,1))

ax1.set_title('LG, 두산, 한화의 연도별 순위')
ax2.set_title('LG, 두산, 한화의 연도별 승률')

ax1.set_ylabel('순위',rotation=0)
ax2.set_ylabel('승률', rotation=0,labelpad=15)

plt.show()
```
