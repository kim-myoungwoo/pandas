```python
import folium
import pandas as pd
import json
```

```python
df=pd.read_excel('./서울지역 대학교 위치.xlsx', engine='openpyxl')

df= df.rename(columns={'Unnamed: 0':'대학명'})
df=df.set_index('대학명')
```

```python
df.columns
```

```python
seoul_map = folium.Map(location=[37.55,126.98],zoom_start=12)
```

```python
for name, lat, lng in zip(df.index, df.위도, df.경도):
    folium.Marker([lat,lng], popup=folium.Popup(name,max_width=1000)).add_to(seoul_map)
    
seoul_map.save('./seoul_collenge.html')
seoul_map
```

```python
for name, lat, lng in zip(df.index, df.위도, df.경도):
    folium.CircleMarker([lat, lng], radius=10, color='brown', fill=True, fill_color='coral', fill_opacity=0.7, popup=name).add_to(seoul_map)
    
seoul_map.save('./seoul+collenge2.html')
    
seoul_map
```

```python
file_path = './경기도인구데이터.xlsx'
df = pd.read_excel(file_path, index_col='구분', engine='openpyxl')
df.columns=df.columns.map(str)

df
```

```python
geo_path = './경기도행정구역경계.json'
try:
    geo_data = json.load(open(geo_path, encoding='utf-8'))
except:
    geo_data = json.load(open(geo_path, encoding='utf-8-sig'))
    
```

```python
g_map = folium.Map(location=[37.5502,126.982],tiles='Stamen Terrain', zoom_start=9)
year ='2007'

folium.Choropleth(geo_data=geo_data,data=df[year],columns=[df.index,df[year]],fill_color='YlOrRd',fill_opacity=0.5,line_opacity=0.3
                 ,threshold_scale=[10000,100000,300000,500000,700000],key_on='feature.properties.name').add_to(g_map)

g_map
```
