```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from matplotlib import font_manager, rc
font_path = "./malgun.ttf"
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)
```

```python
df=pd.read_csv('./auto-mpg.csv',header=None)
```

```python
df.columns = ['mpg','cylinders','displacement','horsepower','weight',
              'acceleration','model year','origin','name']
```

```python
plt.style.use('classic')

df['mpg'].plot(kind='hist',bins=10,color='yellow',figsize=(10,5))

plt.title('Histogram')
plt.xlabel('mpg')
plt.show()
```

```python
plt.style.use('default')

df.plot(kind='scatter',x='weight',y='mpg',c='coral',s=10,figsize=(15,5))
plt.title('Scatter Plot - mpg vs weight')
plt.show()
```

```python
cylinders_size = df.cylinders/df.cylinders.max() * 300

df.plot(kind='scatter',x='weight',y='mpg',c='coral',figsize=(17,10),s=cylinders_size,alpha=0.3)
plt.title('Scatter Plot: mpg-weight-cylinders')

plt.show()
```

```python
cylinders_size = df.cylinders/df.cylinders.max() * 300

df.plot(kind='scatter',x='weight',y='mpg',marker='+',figsize=(13,8),cmap='inferno',c=cylinders_size,s=50,alpha=0.3)
plt.title('Scatter Plot: mpg-weight-cylinders')

plt.savefig("./scatter.png")
plt.savefig('./scatter_transparent.png',transparent=True)

plt.show()
```

```python
df['count']=1
df_origin = df.groupby('origin').sum()
print(df_origin.head())

df_origin.index=['USA','EU','JPN']

df_origin['count'].plot(kind='pie',figsize=(7,5),autopct='%1.1f%%',startangle=10,colors=['chocolate','bisque','cadetblue'])

plt.title('Model Origin',size=20)
plt.axis('equal')
plt.legend(labels=df_origin.index, loc='upper right')
plt.show()
```
